```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = [4, 1, -2, 1, 5, 3]
x2 = [0, 1, 1, 0, 2, -1]
y = [1, -1, 1, 1, -1, -1]
colors = ['r' if label == 1 else 'b' for label in y]

plt.scatter(x1, x2, c=colors)
plt.xlabel("X1")
plt.ylabel("X2")

# Annotate the points with labels from y
for i, label in enumerate(y):
    plt.annotate(str(label), (x1[i] + 0.1, x2[i]))

plt.show()

#(b) From the graph, we can see that the data is NOT linearly seperable because there's no way for us to draw a hyperplane
#perfectly seperating the +1 points from the -1 points. Thus, our algorithm will never converge when we run it 
#several times through the sequence because of the cycling theorem, which tells us that the
#learning algorithm will eventually repeat the same set of parameters and enter an infinite loop
```
